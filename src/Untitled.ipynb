```python
from firecrawl import FirecrawlApp
```

```python
# Initialize the client
firecrawl = FirecrawlApp(api_key="fc-98cff11c9a1d49a0896aa23b1f78ee80")
```

```python
# Define research parameters
params = {
    "maxDepth": 5,  # Number of research iterations
    "timeLimit": 180,  # Time limit in seconds
    "maxUrls": 5  # Maximum URLs to analyze
}
```

```python
# Start research with real-time updates
def on_activity(activity):
    print(f"[{activity['type']}] {activity['message']}")
```

```python
# Run deep research
results = firecrawl.deep_research(
    query="Get the list of real estate in Berlin with price 200000.",
    params=params,
    on_activity=on_activity
)
```
